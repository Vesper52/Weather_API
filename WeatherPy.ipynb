```python
# Dependencies
import openweathermapy.core as ow
import pandas as pd
import random
import time
from citipy import citipy
import matplotlib.pyplot as plt
```

```python
#variables
weather_data = []
cities = []
city_names = []
url = "http://api.openweathermap.org/data/2.5/weather?"
#api_key = "ef11eecce5131794d0534dbed4cfed74"#we changed API keys
#settings = {"units": "imperial", "appid": api_key}

```

```python
api_key = "2d73ec10a02283b78245e5d81a8f5911"#we changed API keys
settings = {"units": "imperial", "appid": api_key}
```

```python
while len(city_names)!=500:
    city=(citipy.nearest_city(random.uniform(-180,180),random.uniform(-90,90)))
    name=city.city_name
    if name  not in (city_names):
        city_names.append(name)
  
    
```

```python
city_list = city_names #storage for current cities for testing
```

```python
for x in range(500):
    sleep = 1
    print ("Processing run number: "+str(x)+" for city "+city_list[x])
    weather_data.append(ow.get_current(city_list[x], **settings))
```

```python
all_names = [data.get('name') for data in weather_data]
```

```python
all_temps = [data.get('main').get('temp') for data in weather_data]
```

```python
all_humidity = [data.get('main').get('humidity') for data in weather_data]
```

```python
all_cloudiness = [data.get('clouds').get('all') for data in weather_data]
```

```python
all_wind_speed = [data.get('wind').get('speed') for data in weather_data]
```

```python
all_latitude = [data.get('coord').get('lat') for data in weather_data]
```

```python
weather_df = {"Cities":all_names,"Temperature (F)":all_temps,"Humidity":all_humidity,"Cloudiness":all_cloudiness,
             "Wind Speed":all_wind_speed,"Latitude":all_latitude}
weather_df= pd.DataFrame(weather_df)
weather_df.head()
```

```python
#
plt.scatter(weather_df["Latitude"], weather_df["Temperature (F)"], marker="o")
plt.title("City Latitude vs Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.grid(True)
plt.savefig("Temperature.png")
plt.show()

#Trend Analysis #1
#What I find interesting here is that when it comes to cities with latitudes near the equator the temperature tends to be very hot 
#and the further you get away from the equator the colder the temperature.
#
#
```

```python
plt.scatter(weather_df["Latitude"], weather_df["Humidity"], marker="o")
plt.title("City Latitude vs Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.savefig("Humidity.png")
plt.show()
#Trend Analysis #2 
#It seems that humidity is all over the place, its hard to get a good grasph on the trend just off of latitude alone like the 
#temperature graph, perhaps if it was a bubble plot with the humidity as the scaling factor it would give a better idea on how
#humidity is based on geographical location.
```

```python
plt.scatter(weather_df["Latitude"], weather_df["Cloudiness"], marker="o")
plt.title("City Latitude vs Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid(True)
plt.savefig("Cloudiness.png")
plt.show()
```

```python
plt.scatter(weather_df["Latitude"], weather_df["Wind Speed"], marker="o")
plt.title("City Latitude vs Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid(True)
plt.savefig("WindSpeed.png")
plt.show()
#Trend Analysis #3
#It seems that no wind conditions are pretty similar across all latitudes around the world, what would be pretty interesting is
#seeing what kind of wind speeds are like in areas affected by hurricanes, but this dataset doesn't really provide that kind of 
#information.
```
